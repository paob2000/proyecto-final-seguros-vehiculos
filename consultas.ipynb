```python
df['Entidad'] = df['Entidad'].astype("string")
df['Municipio'] = df['Municipio'].astype("string")
df['Bien_jurídico_afectado'] = df['Bien_jurídico_afectado'].astype("string")
df['Tipo_de_delito'] = df['Tipo_de_delito'].astype("string")
df['Subtipo_de_delito'] = df['Subtipo_de_delito'].astype("string")
df['Modalidad'] = df['Modalidad'].astype("string")
df['Enero'] = df['Enero'].astype("int64")
df['Febrero'] = df['Febrero'].astype("int64")
df['Marzo'] = df['Marzo'].astype("int64")
df['Abril'] = df['Abril'].astype("int64")
df['Mayo'] = df['Mayo'].astype("int64")
df['Junio'] = df['Junio'].astype("int64")
df['Julio'] = df['Julio'].astype("int64")
df['Agosto'] = df['Agosto'].astype("int64")
df['Septiembre'] = df['Septiembre'].astype("int64")
df['Octubre'] = df['Octubre'].astype("int64")
df['Noviembre'] = df['Noviembre'].astype("int64")
df['Diciembre'] = df['Diciembre'].astype("int64")

df.info()
```

```python
# Número promedio de delitos agrupados por municipio y año-mes de ocurrencia de estos.


# Paso 1: Transformar los datos de columnas de meses a filas (de forma larga)
df_melted = df.melt(
    id_vars=['Año', 'Clave_Ent', 'Entidad', 'Cve_Municipio', 'Municipio', 'Tipo_de_delito', 'Subtipo_de_delito'], 
    value_vars=['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 
                'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    var_name='Mes', 
    value_name='Delitos'
)

# Paso 2: Mapear los meses a números
mes_a_num = {
    'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4, 'Mayo': 5, 'Junio': 6, 
    'Julio': 7, 'Agosto': 8, 'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12
}
df_melted['Mes'] = df_melted['Mes'].map(mes_a_num)

# Paso 3: Crear una columna de "Año-Mes"
df_melted['Año_Mes'] = pd.to_datetime({
    'year': df_melted['Año'], 
    'month': df_melted['Mes'], 
    'day': 1
})

# Paso 4: Agrupar por Municipio y Año-Mes, y calcular el promedio de delitos
resultado = df_melted.groupby(['Municipio', 'Año_Mes'])['Delitos'].mean().reset_index()

# Paso 5: Renombrar las columnas para claridad
resultado.columns = ['Municipio', 'Año_Mes', 'Promedio_Delitos']

# Mostrar el resultado
resultado.sort_values(by='Promedio_Delitos', ascending=False).head(20)
```

```python

# El delito más común por municipio

# Paso 1: Contar la frecuencia de cada tipo de delito por municipio
delito_frecuente = (
    df.groupby(['Municipio', 'Tipo_de_delito'])
    .size()  # Contar la frecuencia
    .reset_index(name='Frecuencia')  # Renombrar la columna del conteo
)

# Paso 2: Identificar el delito más común por municipio
delito_mas_comun = (
    delito_frecuente.loc[
        delito_frecuente.groupby('Municipio')['Frecuencia'].idxmax()
    ]
)

# Paso 3: Renombrar las columnas para claridad
delito_mas_comun.columns = ['Municipio', 'Tipo_de_delito', 'Frecuencia']

# Mostrar el resultado
delito_mas_comun.sort_values(by="Frecuencia", ascending=False).head(20)

```

```python
# El número promedio de delitos agrupados por tipo de delito y año y mes en la ciudad de México ordenados de mayor a menor


# Paso 1: Filtrar los datos para obtener solo los de Ciudad de México
df_cdmx = df_melted[df_melted['Entidad'] == 'Ciudad de México']

# Paso 2: Agrupar por 'Tipo_de_delito' y 'Año_Mes', y calcular el promedio de delitos
promedio_delitos = (
    df_cdmx.groupby(['Tipo_de_delito', 'Año_Mes'])['Delitos']
    .mean()  # Calcular el promedio
    .reset_index(name='Promedio_Delitos')  # Reiniciar el índice
)

# Paso 3: Ordenar de mayor a menor el promedio de delitos
promedio_delitos = promedio_delitos.sort_values(by='Promedio_Delitos', ascending=False)

# Mostrar los resultados
promedio_delitos
```

```python
#  Delitos totales en orden de frecuencia y porcentaje frente al total

import pandas as pd

# Paso 1: Agrupar los datos por 'Tipo_de_delito' y calcular el total de delitos
delitos_totales = (
    df_melted.groupby('Tipo_de_delito')['Delitos']
    .sum()  # Calcular el total de delitos por tipo de delito
    .reset_index()  # Reiniciar el índice
)

# Paso 2: Ordenar los resultados de mayor a menor
delitos_totales = delitos_totales.sort_values(by='Delitos', ascending=False)

# Paso 1: Calcular el total global de delitos
total_delitos = delitos_totales['Delitos'].sum()

# Paso 2: Agregar una columna de porcentaje
delitos_totales['Porcentaje'] = (delitos_totales['Delitos'] / total_delitos) * 100


# Mostrar los resultados
delitos_totales
```
